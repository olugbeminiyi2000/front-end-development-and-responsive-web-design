```python
# with the following function we can select highly correlated columns
#it will remove the first feature
#that is correlated with anything other feature

def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if (abs(corr_matrix.iloc[i, j]) > threshold):
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
                
    return col_corr
# here since both correlated values could cause multicollinearity we
# could either remove the one , on the row by using i or the on on the
# column by using j
```

```python
corr_features = correlation(x, 0.8)
corr_features# the abs help make even the negative and positve correlation 
#positive which will help remove both
#and with or without abs it was still 17
#but it is better to indicate not to indicate abs because negative correlated
#features are also very important for a good module

#dropping the features that can cause multicollinearity
#based on i or j we would switch between both in the model building
```
